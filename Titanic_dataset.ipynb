```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

#data = sns.load_dataset("gender_submission.csv")
data = pd.read_csv("train.csv")
data.head(10)
```

```python
data.columns
```

```python
#survival_rate by passenger class
sns.barplot(data=data,y='Survived',x='Pclass',errorbar=None,palette=['orange','blue','green'])
```

```python
sns.barplot(data=data,x='Survived',y='Pclass')
```

```python
sns.scatterplot(data=data,x='Age',y='Fare',hue='Survived',palette='deep')
```

```python
sns.scatterplot(data=data,x='Age',y='Fare',hue='Sex',palette='deep')
```

```python
#age,fare,pclass
sns.pairplot(data=data, hue='Survived', vars=['Age', 'Fare', 'Pclass'], palette='deep')
```

```python
sns.pairplot(data=data, hue='Sex', vars=['Age', 'Fare', 'Pclass'], palette='deep')
```

```python
selected = data[['Age', 'Fare', 'Pclass','Survived']]
sns.pairplot(data=selected,hue='Survived',palette='deep')
```

```python
data.dtypes
```

```python
correlation_matrix = data.select_dtypes(include='number').corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Tips Dataset')
plt.show()
```
