```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y,color='r', marker='*',linestyle='dotted')  

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=['jan','feb','mar','apr','may','june']
y= [10,20,5,60,22,35]

plt.plot(x,y,color='g',linestyle='dashed',marker='d',linewidth=0.2)
plt.xlabel('months')
plt.ylabel('savings amt')
plt.title('each month savings amount')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=['jan','feb','mar','apr','may','june']
y= [10,20,5,60,22,35]

plt.bar(x,y,color=['brown','orange','lightgreen'],edgecolor='pink',linewidth=3)
plt.xlabel('months')
plt.ylabel('savings amt')
plt.title('each month savings amount')
plt.show()
#ipl scoring,car sales,movie collections
```

```python
import matplotlib.pyplot as plt

y = [10, 20, 30, 40, 50, 60, 60, 70, 80, 90, 100, 10, 20, 30]
plt.hist(y, bins=5, color='g', edgecolor='black')
plt.xlabel('Savings Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Savings Amount')
plt.show()
```

```python
x=[50,10,15,5,20]
y=['webd','java','dsa','apt','js']
plt.pie(x,labels=y, startangle=140)
plt.legend()
```

```python
import matplotlib.pyplot as plt
labels = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
sizes = [50, 25, 10, 10, 5]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'purple']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Programming Languages')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=['jan','feb','mar','apr','may','june']
y= [10,20,5,60,22,35]

plt.scatter(x,y,color='r',edgecolor='white',linewidth=30)
plt.xlabel('months')
plt.ylabel('savings amt')
plt.title('Scatter plot for each month savings amount')
plt.show()
```
